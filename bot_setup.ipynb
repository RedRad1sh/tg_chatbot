```python
vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1,3))
X = vectorizer.fit_transform(X_text)
```

```python
clf = SGDClassifier()
clf.fit(X,y)
clf = CalibratedClassifierCV(clf, cv='prefit')
clf.fit(X,y)
```

```python
def GetIntent(text):
    probas = clf.predict_proba(vectorizer.transform([text]))[0]
    proba = max(probas)
    best_answer = list(probas).index(proba)
    if proba > 0.25:
        return clf.classes_[best_answer]
```
